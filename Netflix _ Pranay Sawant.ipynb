####  TV Shows released in 2021

```python
df.query('Release_year==2021 & Type == "TV Show" ')
```

#### Top 5 countries which relesed TV Shows in 2021

```python
df.query('Release_year==2021 & Type == "TV Show" ')["Country"].value_counts().head()
```

#### TV Shows released by Netflix India in 2021

```python
df.query('Release_year==2021 & Type == "TV Show" & Country == "India" ')
```

#### Movies Released in 2021

```python
df.query('Release_year==2021 & Type == "Movie" ')
```

#### Top 5 countries in releasing movies in 2021
